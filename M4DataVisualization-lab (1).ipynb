## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# your code goes here
Query = """ SELECT ConvertedComp from master"""
df = pd.read_sql_query(Query,conn)

df.dropna(subset=['ConvertedComp'],axis=0, inplace =True)
count, bin_edges = np.histogram(df)
df.plot(kind='hist', figsize = (12,5), xticks = bin_edges)

plt.title('Histogram of Annual Salary in USD')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual Salary')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
Query = """SELECT Age from master"""
df = pd.read_sql_query(Query,conn)

df.plot(kind='box', figsize = (5,8))
plt.title('Plot Showing Age of Developers')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
Query = """SELECT Age, WorkWeekHrs from master"""
df = pd.read_sql_query(Query,conn)

df.plot(kind = 'scatter' , x = 'Age', y = 'WorkWeekHrs', figsize = (12,5))
plt.title("Developers' Age and Weekly Working Hrs")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# your code goes here
Query = """ SELECT WorkWeekHrs, CodeRevHrs, Age from master"""
df = pd.read_sql_query(Query,conn)

df.dropna(subset=['WorkWeekHrs','CodeRevHrs','Age'],inplace=True)

fig = px.scatter(df,x=['WorkWeekHrs','CodeRevHrs'],y='Age', size = 'Age',
                 hover_name = 'Age',title='CoderevHrs and WorkWeekHrs vs Age', size_max = 60)
fig.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query = """SELECT count(*) as Num_of_Respondent,DatabaseDesireNextYear from DatabaseDesireNextYear group by DatabaseDesireNextYear order by count(*) desc limit 5"""
df = pd.read_sql_query(Query,conn)

df.head(5)
```

```python
values = df['Num_of_Respondent']
names = df['DatabaseDesireNextYear']
fig = px.pie(df,values=values, names=names, title='Top 5 databases Developers wish to learn')
fig.show()
```

```python
Query = """select LanguageDesireNextYear,count(*) as Num_of_Respondents from LanguageDesireNextYear group by LanguageDesireNextYear order by Num_of_Respondents desc"""
df = pd.read_sql_query(Query,conn)
df
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query = """SELECT WorkWeekHrs, CodeRevHrs,Age from master where Age between 30 and 35"""
df = pd.read_sql_query(Query,conn)

df1 = df.groupby('Age').median()
df1
```

```python
# your code goes here
Query = """SELECT WorkWeekHrs, CodeRevHrs, Age from master where Age between 30 and 35"""
df = pd.read_sql_query(Query,conn)

df1 = df.groupby('Age').median()
df1.plot(kind='area',
         stacked=True,
         figsize = (10,5))
plt.title('Median WorkWeekHrs and CodeRevHrs for the group 30 to 35')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
Query = """SELECT ConvertedComp, Age from master where Age between 45 and 60"""
df= pd.read_sql_query(Query,conn)
df1 = df.groupby('Age').median()
df1.reset_index()
```

```python
# your code goes here
Query = """SELECT ConvertedComp, Age from master where Age between 45 and 60"""
df= pd.read_sql_query(Query,conn)

df1 = df.groupby('Age').median()
df1.plot(kind='line',figsize=(10,5))
plt.title('Annual Salary for all Ages between 45 and 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```
